```python
x_train=np.array([1.0,2.0])
y_train=np.array([300.0,500.0])
```

```python
def compute_cost(x,y,w,b):
    m=x.shape[0]
    cost=0
    for i in range(m):
        f_wb=w*x[i]+b
        cost=cost+(f_wb-y[i])**2
    total_cost=(1/(2*m))*cost
    return total_cost
```

```python
def compute_gradient(x,y,w,b):
    """
    Computes gradient for linear regression
    Args:
        x (ndarray(m,)): Data, m examples
        y (ndarray(m,)): target values
        w,b (scalars): model parameters
    Returns
        dj_dw (scalar): The gradient of the cost w.r.t the parameter w
        dj_db (scalar): The gradient of the cost w.r.t the parameter b
    """

    m=x.shape[0]
    dj_dw=0
    dj_db=0

    for i in range(m):
        f_wb=w*x[i]+b
        dj_dw_i=(f_wb-y[i])*x[i]
        dj_db_i=(f_wb-y[i])
        dj_dw+=dj_dw_i
        dj_db+=dj_db_i
    dj_dw=dj_dw/m
    dj_db=dj_db/m
    
    return dj_dw,dj_db
```

```python
plt_gradients(x_train,y_train,compute_cost,compute_gradient)
plt.show()
```

```python
def gradient_descent(x,y,w_in,b_in,alpha,num_iters,cost_function,gradient_function):
    """
    Performs gradient descent to fit w,b. Updates w,b by taking num_iters gradient steps with learning rate alpha

    Args:
        x (ndarray(m,)): Data, m examples
        y (ndarray(m,)): target values
        w_in, b_in: initial values of model parameters
        alpha (float): learning rate
        num_iters (int): number of iterations to run gradient descent
        cost_function: function to call to produce cost
        gradient_function: function to call to produce gradient
    
    Returns:
        w (scalar): Updated value of parameter after running gradient descent
        b (scalar): Updated value of parameter after running gradient descent
        J_history (List): History of cost values
        p_history (List): History of parameters [w,b]
    """

    w=copy.deepcopy(w_in)
    J_history=[]
    p_history=[]
    b=b_in
    w=w_in

    for i in range(num_iters):
        dj_dw,dj_db=gradient_function(x,y,w,b)
        b=b-alpha*dj_db
        w=w-alpha*dj_dw

        if i<100000:
            J_history.append(cost_function(x,y,w,b))
            p_history.append([w,b])
        if i% math.ceil(num_iters/10)==0:
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e}  ",
                  f"dj_dw: {dj_dw:0.3e}, dj_db: {dj_db:0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")
    return w,b, J_history,p_history
```

```python
w_init=0
b_init=0
iterations=10000
tmp_alpha=0.01
w_final,b_final,J_hist,p_hist=gradient_descent(x_train,y_train,w_init,b_init,tmp_alpha,iterations,compute_cost,compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```

```python
w_init=0
b_init=0
iterations=10000
tmp_alpha=0.01

def gradient_descent_1(x,y,w_init,b_init,alpha,num_iter):
    m=x.shape[0]
    w=w_init
    b=b_init
    for i in range(num_iter):
        dj_dw=0
        dj_db=0
        for j in range(m):
            f_wb=w*x[j]+b
            dj_dw_j=(1/m)*(f_wb-y[j])*x[j]
            dj_db_j=(1/m)*(f_wb-y[j])
            dj_dw+=dj_dw_j
            dj_db+=dj_db_j
        w=w-alpha*dj_dw
        b=b-alpha*dj_db
    return w,b

gradient_descent_1(x=x_train,y=y_train,w_init=w_init,b_init=b_init,alpha=tmp_alpha,num_iter=iterations)
```
